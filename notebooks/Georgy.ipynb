```python
import os
import pandas as pd

df = pd.read_csv('/Users/maksim/code/georgylevin27/ValTool/data/Data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df and the relevant columns are named 'Revenue' and 'EV'
plt.figure(figsize=(10, 6))
plt.scatter(df['Target_Revenues'], df['Target_EV'], color='blue', alpha=0.5)

# Set the labels and title
plt.xlabel('Revenue')
plt.ylabel('Enterprise Value (EV)')
plt.title('Scatter Plot of Enterprise Value vs Revenue')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Scatter plot with smaller points and higher transparency
plt.figure(figsize=(10, 6))
plt.scatter(df['Target_Revenues'], df['Target_EV'], s=1, alpha=0.3, color='blue')  # 's' controls point size
plt.xlabel('Revenue')
plt.ylabel('Enterprise Value (EV)')
plt.title('Scatter Plot of Enterprise Value vs Revenue')
plt.show()

# Scatter plot with logarithmic scale
plt.figure(figsize=(10, 6))
plt.scatter(df['Target_Revenues'], df['Target_EV'], s=1, alpha=0.3, color='blue')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Revenue (Log Scale)')
plt.ylabel('Enterprise Value (EV) (Log Scale)')
plt.title('Scatter Plot of Enterprise Value vs Revenue (Log Scale)')
plt.show()

# Hexbin plot
plt.figure(figsize=(10, 6))
plt.hexbin(df['Target_Revenues'], df['Target_EV'], gridsize=50, cmap='Blues', mincnt=1)
plt.colorbar(label='Counts')
plt.xlabel('Revenue')
plt.ylabel('Enterprise Value (EV)')
plt.title('Hexbin Plot of Enterprise Value vs Revenue')
plt.show()
```
